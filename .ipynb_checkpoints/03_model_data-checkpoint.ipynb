```python
import mysql.connector
from mysql.connector import Error
import pandas as pd
import MySQLdb
```

```python
#Conexion a la base de datos de mysql proyecto_final
mysql_conexion =  MySQLdb.connect(host = 'localhost',
                                 port = 3306,
                                 user = 'root',
                                 passwd = 'Adriff22',
                                 db = 'proyecto_final')
```

```python
#Vamos a hacer una consulta a la data del DENUE para trabajarlo como un dataframe
denue_cdmx_713943 = pd.read_sql('select * from dnue_cdmx_713943;', mysql_conexion) 
denue_cdmx_713943.head(5)
```

```python
#Ahora la data de los Censos Económicos anivel de municipio 
ce_cdmx_mpio_713943 = pd.read_sql('select * from ce_cdmx_mpio_713943_fin;', mysql_conexion) 
ce_cdmx_mpio_713943.head(5)
```

```python
#Ahora la data de los Censos Económicos anivel de estrato de personal ocupado 
ce_cdmx_espo_713943 = pd.read_sql('select * from ce_cdmx_espo_713943_fin;', mysql_conexion) 
ce_cdmx_espo_713943.head(5)
```

```python
#El objetivo será distribuir los valores de los indicadores del df ce_cdmx_mpio_713943 
#para cada unidad económica de denue_cdmx_713943
#Primero vamos a crear un df agrupando la data del dnue de municipios y estrato

denue_cdmx_mpio_espo = denue_cdmx_713943[[
                                       'actividad_economica',                                        
                                       'entidad_federativa', 
                                       'municipio',
                                       'estrato'
                                      ]]
```

```python
denue_cdmx_mpio_espo.head()
```

```python
denue_01 = pd.DataFrame(
                        denue_cdmx_mpio_espo
                        .assign(mpio_espo = lambda x: x['municipio'] + '_' + x['estrato'])
                        )
denue_01.head()
```

```python
#Aquí vamos a obtener un data frame que nos de el número de establecimientos por municipio y estrato
denue_02 = pd.DataFrame(
                        denue_01
                        .assign(mpio_espo_n = 1)
                        .groupby(['mpio_espo', 'municipio','estrato'])['mpio_espo_n']
                        .sum()
                        .reset_index()
                        )
denue_02.head()
```

```python
#Ahora creamos otro df con el conteo de UE del sector por municipio
denue_03 = pd.DataFrame(
                        denue_cdmx_mpio_espo
                        .assign(municipio_n = 1)
                        .groupby(['municipio'])[ 'municipio_n']
                        .sum()
                        .reset_index()
                        )
denue_03.head()
```

```python
#Hacemos un merge de denue_02 y denue_03
denue_04 = denue_02.merge(denue_03, on = 'municipio', how = 'left')
denue_04.head()
```

```python
#En este df calculamos la indice para cada estrato y su respectivo municipio, que servirá para distribuir el valor del indicador
denue_05 = pd.DataFrame(
                        denue_04
                        .assign(u = lambda x: x['mpio_espo_n'] / x['municipio_n'])
                        )
denue_05.head()
```

```python
#Vamos a construir un df con los valores de los indicadores pero solo a nivel de municipio ce_cdmx_mpio_713943

ce_mpio = pd.DataFrame(
                        ce_cdmx_mpio_713943
                        .groupby('municipio')[['H001A', 'H001B', 'H001C', 'K000A', 'A700A', 'J000A', 'M000A', 'A800A']]
                        .sum()
                        .reset_index()
                        )
ce_mpio.head()
```

```python
#Ahora vamoa hacer un merge con denue_05
denue_06 = denue_05.merge(ce_mpio, on = 'municipio', how = 'left')
denue_06.head()
```

```python
#Esta operación permite calular la proporción que le toca a cada establecimiento, según su municipio y
#su estrato de personal ocupado.
denue_07 = pd.DataFrame(
                        denue_06
                        .assign(H001A_mpio = lambda x: x['u'] * x['H001A'] / x['mpio_espo_n'])
                        .assign(H001B_mpio = lambda x: x['u'] * x['H001B'] / x['mpio_espo_n'])
                        .assign(H001C_mpio = lambda x: x['u'] * x['H001C'] / x['mpio_espo_n'])
                        .assign(K000A_mpio = lambda x: x['u'] * x['K000A'] / x['mpio_espo_n'])
                        .assign(A700A_mpio = lambda x: x['u'] * x['A700A'] / x['mpio_espo_n'])
                        .assign(J000A_mpio = lambda x: x['u'] * x['J000A'] / x['mpio_espo_n'])
                        .assign(M000A_mpio = lambda x: x['u'] * x['M000A'] / x['mpio_espo_n'])
                        .assign(A800A_mpio = lambda x: x['u'] * x['A800A'] / x['mpio_espo_n'])
                        )
denue_07.head()
```

```python
#Nosquedamos solo con los campos de interés

denue_ce_mpio = denue_07[['mpio_espo', 'mpio_espo_n',
                          'H001A_mpio','H001B_mpio','H001C_mpio','K000A_mpio',
                          'A700A_mpio','J000A_mpio','M000A_mpio','A800A_mpio']]
denue_ce_mpio.head()
```

```python
#Ahora vamos a distribuir el total de cada indicador por establecimiento, según su estrato de peronal ocupado

denue_ce_establ = (denue_cdmx_713943[[
                                     'actividad_economica',
                                     'entidad_federativa', 
                                     'municipio',
                                     'cp',
                                     'calle',
                                     'colonia',
                                     'num_exterior',
                                     'num_interior',
                                     'telefono',
                                     'sitio_internet',
                                     'correo_e',
                                     'nombre',
                                     'razon_social',
                                     'latitud',
                                     'longitud',
                                     'estrato']].assign(mpio_espo = lambda x: x['municipio'] + '_' + x['estrato']))
                                     

denue_ce_establ.head()
```

```python
#Ahora vamos a cruzar el df de establecimientos denue_ce_establ con denue_ce_mpio, del cual
#resultará nuestro modelo.

denue_ce_model =  (
                    denue_ce_establ
                    .merge(denue_ce_mpio, on = 'mpio_espo', how = 'left')
                  )
```

```python
#Ahora vamos a indexar nuestra data resultado del modelo a MySQL

from sqlalchemy import create_engine
engine = create_engine("mysql+mysqldb://root:"+'Adriff22'+"@localhost/proyecto_final")
```

```python
denue_ce_model.to_sql("denue_ce_model_713943", con = engine, if_exists = 'replace', index = True)
```
